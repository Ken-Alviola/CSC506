```python
import random
numbers = random.sample(range(500),500)
print('NUMBERS:', numbers)
```

```python
numbers_original = numbers
```

```python
def selection_sort(numbers):
    # A variable to hold the number of item comparisons
    comparisons = 0
    
    for i in range(len(numbers)-1):
        # Find index of smallest remaining element
        index_smallest = i
        for j in range(i+1, len(numbers)):
            comparisons = comparisons + 1
            if numbers[j] < numbers[index_smallest]:
                index_smallest = j

        # Swap numbers[i] and numbers[index_smallest]
        temp = numbers[i]
        numbers[i] = numbers[index_smallest]
        numbers[index_smallest] = temp

    return comparisons
```

```python
# Main program to test the selection sort algorithm
# Display the contents of the list
print('UNSORTED:', numbers)
```

```python
# Call the selection_sort() function
comparisons = selection_sort(numbers)
```

```python
# Display the (sorted) contents of the list
print('Total comparisons: %d' % comparisons)
print('SORTED:', numbers)
```

```python
def insertion_sort(num_list):
    for i in range(1, len(num_list)):
        j = i

        # Insert num_list[i] into sorted part 
        # stopping once num_list[i] in correct position
        while j > 0 and num_list[j] < num_list[j - 1]:
            # Swap num_list[j] and num_list[j - 1]
            temp = num_list[j]
            num_list[j] = num_list[j - 1]
            num_list[j - 1] = temp
            j -= 1
```

```python
# Create a list of unsorted values    
numbers = numbers_original
```

```python
# Print unsorted list
print('UNSORTED:', numbers)
```

```python
# Initial call to insertion_sort with index 
insertion_sort(numbers)
```

```python
# Print sorted list
print('SORTED:', numbers)
```
